# PCA, t-SNE

PCA and t-SNE are used to project data from high dimensions down to lower dimensions based on different objectives.

PCA: Summarize data using linear transformations (preserve the maximum amount of variance)

t-SNE: Try to preserve distances between nearby points (large distances don't matter as much).

```python
import numpy as np
import csv

rows = []

with open('Fish.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        rows.append(row)

print(len(rows))
print(rows[0]) # first row is a header
print(rows[1])

rows = rows[1:]

labels = {} # Create a dictionary of label strings to numeric values
for row in rows:
    if row[0] not in labels:
        labels[row[0]]=len(labels)

print(labels)
        
inputs = np.array([row[1:] for row in rows]).astype(np.float32)
outputs = np.array([labels[row[0]] for row in rows])
print(outputs)
```

```python
from sklearn.decomposition import PCA
X_embedded = PCA(n_components=2).fit_transform(inputs)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.scatter(X_embedded[:, 0], X_embedded[:,1], c=outputs)
plt.show()
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(inputs)
```

```python
plt.scatter(X_embedded[:, 0], X_embedded[:,1], c=outputs)
plt.show()
```
